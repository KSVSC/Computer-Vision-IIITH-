```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os

# Load an image using opencv
img1 = cv2.imread("../images/girl.png")
img2 = cv2.imread("../images/pigeon.jpg")
img3 = cv2.imread("../images/unsplash.jpg")
img4 = cv2.imread("../images/virus.jpg")
```

```python
# Display the images using matplotlib 
plt.figure()
f,ax = plt.subplots(2,2)
ax[0,0].imshow(img1)
ax[0,1].imshow(img2)
ax[1,0].imshow(img3)
ax[1,1].imshow(img4)
```

```python
# Correct the color of images (RGB->BGR)
plt.figure()
f,ax = plt.subplots(2,2)
ax[0,0].imshow(cv2.cvtColor(img1,cv2.COLOR_BGR2RGB))
ax[0,1].imshow(cv2.cvtColor(img2,cv2.COLOR_BGR2RGB))
ax[1,0].imshow(cv2.cvtColor(img3,cv2.COLOR_BGR2RGB))
ax[1,1].imshow(cv2.cvtColor(img4,cv2.COLOR_BGR2RGB))
```

```python
#Convert any colour image (of your choice) to grayscale without using a function
img1_rgb = img1[:, :, ::-1]
img2_rgb = img2[:, :, ::-1]
img3_rgb = img3[:, :, ::-1]
img4_rgb = img4[:, :, ::-1]
gray_img1 = (img1_rgb[:,:,0] + img1_rgb[:,:,1] + img1_rgb[:,:,2]) / 3
gray_img1 = np.mean(img1_rgb, axis=2)
gray_img2 = np.mean(img2_rgb, axis=2)
gray_img3 = np.mean(img3_rgb, axis=2)
gray_img4 = np.mean(img4_rgb, axis=2)

path = '../images/'
cv2.imwrite(os.path.join(path , "girl_gray.png"), gray_img1)
cv2.imwrite(os.path.join(path , "pigeon_gray.jpg"), gray_img2)
cv2.imwrite(os.path.join(path , "specs_gray.jpg"), gray_img3)
cv2.imwrite(os.path.join(path , "virus_gray.jpg"), gray_img4)
```
